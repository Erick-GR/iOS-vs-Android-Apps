```python
import pandas as pd
import time
```

```python
# Read Google Playstore csv

fn = "google-playstore-apps/Google-Playstore-Full.csv"

google = pd.read_csv(fn)

google.head()
```

```python
# Delete extra columns

google = google.drop(['Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14'], axis = 1)
google.head()
```

# Perform a series of groupby as to find out what columns or rows to clean.

```python
## Check price consistency

g_price = google.groupby(["Price"])
g_price.count()
```

```python
# Delete rows with erroneous data

toDrop = list(g_price.indices.keys())
toKeep = toDrop[:-16]

google = google[google['Price'].isin(toKeep)]
google['Price'].unique()
```

```python
google_category = google.groupby(["Category"])
google_category.count()
```

```python
# Data is clean

google.Category.unique()
```

```python
google_engine = create_engine(f'postgresql://postgres:postgres@localhost:5433/{db}')

google_conn = google_engine.connect()
meta = MetaData()

for c in cat:
    
    globals()[c+"_tbl"] = Table(
        c+'_tbl', meta,
        Column('id', Integer, primary_key=True),
        Column('App Name', String),
        Column('Category', String),
        Column('Rating', String),
        Column('Reviews', String),
        Column('Installs', String),
        Column('Size', String),
        Column('Price', String),
        Column('Content Rating', String),
        Column('Last Updated', String),
        Column('Minimum Version', String),
        Column('Latest Version', String)
    )
    
meta.create_all(google_engine)


#      premise_transformed.to_sql(name='premise', con=engine, if_exists='append', index=True)
```
